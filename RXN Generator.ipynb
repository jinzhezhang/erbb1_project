```python
import rdkit.Chem as Chem
from rdkit.Chem import Draw
from rdkit.Chem import Descriptors
from rdkit.Chem import Draw
from cairosvg import svg2png
```

```python
filenames = ['erbb18_final.txt']
all_smiles =[]
all_scores =[]
all_reactions=[]
for filename in filenames:
    with open("rxngenerator/Results/"+filename) as reader:
        lines = reader.readlines()
        for line in lines:
            line = line.strip()
            res = line.split(" ")
            smiles = res[0]
            score = float(res[-1])
            reactions = res[1:-1]
            all_smiles.append(smiles)
            all_scores.append(score)
            all_reactions.append(reactions)


# filter
b_all_smiles =[]
b_all_scores =[]
b_all_reactions=[]
for i in range(len(all_smiles)):
    if all_smiles[i] not in b_all_smiles:
        b_all_smiles.append(all_smiles[i])
        b_all_reactions.append(all_reactions[i])
        b_all_scores.append(-all_scores[i])
pairs = [(smiles, rxn, score) for smiles, rxn, score in zip(b_all_smiles, b_all_reactions, b_all_scores)]
pairs = sorted(pairs, key=lambda x:x[2], reverse=False)

mols = [Chem.MolFromSmiles(s) for s, _, _ in pairs[:50]]
vals = ["%.4f" % score for _,_,score in pairs[:50]]
#print(mols)
Draw.MolsToGridImage(mols, molsPerRow=5, subImgSize=(200,135), legends=vals, useSVG=True)
```

# Visualize USPTO Dataset erbb1 activity distribution

```python
%cd $cwd
from matplotlib import pyplot as plt
import matplotlib.font_manager
plt.rcParams['font.family'] = 'sans-serif'
plt.rcParams['font.sans-serif'] = 'Helvetica'
plt.rcParams['axes.edgecolor']='#333F4B'
plt.rcParams['axes.linewidth']=0.8
plt.rcParams['xtick.color']='#333F4B'
plt.rcParams['ytick.color']='#333F4B'
```

```python
chembl_df = pd.read_csv('rxngenerator/erbb1_log_50.csv', sep = ',', names = ['smiles', 'log_ic50'])

file = 'rxngenerator/USPTO_res.txt'
plt.figure(figsize=(8, 6), dpi=120)
smiles_dict = {}
score_list = []
unique = 0
valid = 0
with open(file, 'r') as f:
    for line in f.readlines():
        if '|' in line:
            smiles, score = line.split('|')
            if smiles not in smiles_dict:
                smiles_dict[smiles] = score
                unique += 1
            if "None" not in score:
                valid += 1
                score_list.append(float(score))
with open(file, 'r') as f:
    for line in f.readlines():
        if '|' in line:
            smiles, score = line.split('|')
            if smiles not in smiles_dict:
                smiles_dict[smiles] = score
                unique += 1
            if "None" not in score:
                valid += 1
                score_list.append(float(score))
plt.hist(score_list, bins = 100, color = '#3498DB', alpha = 0.5, label = 'USPTO')
plt.hist(chembl_df['log_ic50'],bins = 100, color = '#FA8072', alpha = 0.5, label = 'ChEMBL')
plt.hist(b_all_scores * 2, bins = 20, color = '#1ABC9C', alpha = 0.5, label = 'RXN Generator')
plt.legend(loc = 'best')
plt.xlabel('log_ic50')
plt.ylabel('frequnce')
plt.title('erbb1 distribution comparison between ChEMBL training set and USPTO')
plt.show()
print(unique, valid)
```

What we can observe from the BO result is that the generated best molecules are has a score which is not low enough comparing to other molecule generator's result.

To verify if this phenomenon is due to some implementation error, we decided to train a traditional JT-VAE with the same implementation. Please refer to the JT-VAE ipython notebook for details.

As a summary, JT-VAE produces similarly bad result (high ic50 value even after BO) if we use a random dataset to train VAE. However, if we retrain the JT-VAE using the erbb1 smiles dataset (erbb1_ligand.csv) and proceed with the same code, JT-VAE can produce good result. 

We believe this faillure is due to a in-effect double-extrapolation mecanisme: 

The BO is trained only on USPTO molecules, which are mostly outside of erbb1 dataset, this is the first extrapolation:  we are training erbb1 activity value over latent space using extrapolation of GNN predictive model. 

Then, the reverse-extrapolation: BO will predict promising points over latent space, which are oftent outside of its training set (USPTO). Although the training and testing MAE for BO is low, we have no measurement in place to evaluate if BO predicted good points are also good depends on our GNN predictive model. 


Due to this double-extrapolation mecanisme, BO is working on a learned space that lacks of accuracy. So in return, BO predicted good points seems like random sampling from latent space.

# Perspective

In this study, although we are unable to generate molecules which has extremly low log ic50 value, we have sucessfully generated several good candidates with their synthetic path way planned.

We believe this is a benefical approach for our clients in phamaceutical industry and the problem we are facing (ineffective BO) can be solved with more developement time given.

Our plan to be solve ineffective BO issue can be done using the following strategy (which we tested over JT-VAE[1] and the result has proven similar strategy is effective):

- Retrain the latent space over erbb1 dataset, or any dataset that is aligned with the training set of our GNN predictive model. In this way, we avoid the double-extrapolation situation in the concept by aligning latent space to our labeled dataset. Doing this is easy for traditional VAE, however, in the case of RXN Generator[2], we need to provide full synthetic pathway for each signle training molecule. This later issue can be solved using current multi-step synthetic pathway planning methods, such as Molecular Transformer[3]. 

For future improvement, we believe this method can be combined with [4] for a more effective optimization over latent space.

[1]. Jin, Wengong, Regina Barzilay, and Tommi Jaakkola. "Junction tree variational autoencoder for molecular graph generation." International conference on machine learning. PMLR, 2018.

[2]. Nguyen, Dai Hai, and Koji Tsuda. "A generative model for molecule generation based on chemical reaction trees." arXiv preprint arXiv:2106.03394 (2021).

[3]. Schwaller, Philippe, et al. "Molecular transformer: a model for uncertainty-calibrated chemical reaction prediction." ACS central science 5.9 (2019): 1572-1583.

[4]. Tripp, Austin, Erik Daxberger, and Jos Miguel Hernndez-Lobato. "Sample-efficient optimization in the latent space of deep generative models via weighted retraining." Advances in Neural Information Processing Systems 33 (2020).
